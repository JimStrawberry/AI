```python
import pandas as pd

url = "https://raw.githubusercontent.com/JimStrawberry/AI/refs/heads/master/ML/dataset/Salary_Data2.csv"
data = pd.read_csv(url)
data
```

```python
data["EducationLevel"] = data["EducationLevel"].map({"高中以下": 0, "大學": 1, "碩士以上": 2})
data
```

```python
from sklearn.preprocessing import OneHotEncoder

onehotencoder = OneHotEncoder()
city_encoded = onehotencoder.fit_transform(data[["City"]]).toarray()
city_encoded
```

```python
data[["CityA", "CityB", "CityC"]] = city_encoded
data
```

```python
data.drop(["CityC", "City"], axis=1, inplace=True)
data
```

```python
from sklearn.model_selection import train_test_split

x = data[["YearsExperience", "EducationLevel", "CityA", "CityB"]]
y = data["Salary"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=23)
len(x_train), len(x_test)
x_train.to_numpy()
x_test.to_numpy()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
x_train
```

```python
import numpy as np

w = np.array([1,2,3,4])
b = 1
y_pred = (x_train*w).sum(axis=1) + b
y_pred
```

```python
((y_train - y_pred)**2).mean()
```

```python
def compute_cost(x, y, w, b):
    y_pred = (x*w).sum(axis=1) + b
    cost = ((y - y_pred)**2).mean()
    return cost
```

```python
w = np.array([1,2,3,4])
b = 1
cost = compute_cost(x_train, y_train, w, b)
cost
```

```python
y_pred = (x_train*w).sum(axis=1) + b
b_gradient = (y_pred - y_train).mean()
w_gradient = np.zeros(x_train.shape[1])

for i in range(x_train.shape[1]):
    w_gradient[i] = (x_train[:, i]*(y_pred - y_train)).mean()

w_gradient, b_gradient
```

```python
def compute_gradient(x, y, w, b):
  y_pred = (x*w).sum(axis=1) + b
  b_gradient = (y_pred - y).mean()
  w_gradient = np.zeros(x.shape[1])

  for i in range(x.shape[1]):
      w_gradient[i] = (x[:, i]*(y_pred - y)).mean()
  return w_gradient, b_gradient

w_gradient, b_gradient = compute_gradient(x_train, y_train, w, b)
w_gradient, b_gradient
```

```python
learning_rate = 0.001
w_gradient, b_gradient = compute_gradient(x_train, y_train, w, b)
print(compute_cost(x_train, y_train, w, b))
w = w - w_gradient*learning_rate
b = b - b_gradient*learning_rate
w, b
print(compute_cost(x_train, y_train, w, b))
```

```python
np.set_printoptions(formatter={'float': '{: .2e}'.format})
def gradient_descent(x, y, w_init, b_init, cost_function, gradient_function, learning_rate, iterations, p_iter=1000):
  c_hist = []
  w_hist = []
  b_hist = []

  w = w_init
  b = b_init

  for i in range(iterations):
    w_gradient, b_gradient = gradient_function(x, y, w, b)

    w = w - w_gradient*learning_rate
    b = b - b_gradient*learning_rate
    cost = cost_function(x, y, w, b)

    c_hist.append(cost)
    w_hist.append(w)
    b_hist.append(b)


    if i%p_iter == 0:
      print(f"Iteration {i:5} : Cost {cost: .4e}, w: {w}, b: {b: .2e}, w_gradient: {w_gradient}, b_gradient: {b_gradient: .2e}")

  return w, b, w_hist, b_hist, c_hist
```

```python
w_init = np.array([1, 2, 3, 4])
b_init = 0
learning_rate = 1.0e-2
run_iter = 10000

w_final, b_final, w_hist, b_hist, c_hist = gradient_descent(x_train, y_train, w_init, b_init, compute_cost, compute_gradient, learning_rate, run_iter)
```

```python
w_final, b_final
```

```python
y_pred = (w_final*x_test).sum(axis=1) + b_final
pd.DataFrame({
    "y_pred": y_pred,
    "y_test": y_test
})
```
