```python
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

# 1.get iris dataset
iris = load_iris()
# feature name
print(iris.feature_names)
```

```python
# 2.dara pre-processing
# 2.1 data split
x_train,x_test,y_train,y_test = train_test_split(iris.data,iris.target,test_size=0.3,random_state=2)
# 2.2 standardization
transfer = StandardScaler()
x_train = transfer.fit_transform(x_train)
x_test = transfer.fit_transform(x_test)
x_train
```

```python
# 3.model training
# 3.1 Instantiate the classifier
estimator = KNeighborsClassifier(n_neighbors=9)

# 3.2 Using Cross-Validation Grid Search
'''
n_neighbors can be one of 1, 3, 5, 7, 9, 11, and GridSearchCV will
automatically test these different values ​​to find the optimal n_neighbors value.
'''
params_grid = {"n_neighbors":[1,3,5,7,9,11]}

'''
GridSearchCV is a tool provided by scikit-learn for automated hyperparameter
tuning. It uses the specified parameter grid (params_grid) to test different
parameter combinations and evaluates the effect of each combination through
cross-validation (cv=5). This step will perform 5 cross-validations for each
n_neighbors value to ensure more stable results, and then select the
optimal parameter combination.
'''
estimator = GridSearchCV(estimator,param_grid=params_grid,cv=5)

# 3.3 model training
estimator.fit(x_train,y_train)
```

```python
# 4.model evaluation
y_pre = estimator.predict(x_test)
print(y_pre)
print(y_test)
```

```python
# accuracy
ret = estimator.score(x_test,y_test)
print("accuracy:",ret)
```

```python
print('The best model:',estimator.best_estimator_)
print('The best score:',estimator.best_score_)
print('The best result:',estimator.cv_results_)
```
